## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
import pandas as pd
new_df = pd.read_csv('Resources/client_dataset.csv')
new_df['sub_total'] = new_df['unit_price'] * new_df['qty']
new_df.head()


```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

# create a function to determine the shipping price
def shipping_price(weight):
    if weight > 50:
        return weight * 7
    else:
        return weight * 10
    
new_df['shipping_price'] =   new_df['unit_weight'].apply(shipping_price)
new_df.head()

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

new_df['total_price'] = (new_df['sub_total']+ new_df['shipping_price']) * 1.0925
new_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
new_df['line_cost'] = new_df['unit_cost'] * new_df['qty'] + new_df['shipping_price']
new_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price

new_df['line_profit'] = new_df['total_price'] - new_df['line_cost']
new_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
top_clients_spending = new_df[new_df['client_id'].isin(top_5_clients)].groupby('client_id')['total_price'].sum()
# Convert the result to a DataFrame for better display
top_clients_spending_df = top_clients_spending.reset_index().rename(columns={'client_id': 'Client ID', 'total_price': 'Total Spending'})

# Format the DataFrame with commas
top_clients_spending_df['Total Spending'] = top_clients_spending_df['Total Spending'].apply(lambda x: f"${x:,.2f}")
top_clients_spending_df
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_clients_df = new_df[new_df['client_id'].isin(top_5_clients)]
# Create the summary DataFrame
summary_df = top_clients_df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'line_profit': 'sum'
}).reset_index()

# Rename columns for clarity
summary_df.columns = ['Client ID', 'Total Units Purchased', 'Total Shipping Price', 'Total Revenue', 'Total Profit']

# Format the DataFrame for better readability
summary_df['Total Revenue'] = summary_df['Total Revenue'].apply(lambda x: f"${x:,.2f}")
summary_df['Total Shipping Price'] = summary_df['Total Shipping Price'].apply(lambda x: f"${x:,.2f}")
summary_df['Total Profit'] = summary_df['Total Profit'].apply(lambda x: f"${x:,.2f}")
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 


# Define a function that converts a dollar amount to millions.


# Apply the currency_format_millions function to only the money columns. 


# Rename the columns to reflect the change in the money format. 

# Filter data for the top 5 clients
top_clients_data = new_df[new_df['client_id'].isin(top_5_clients)]

# Create the summary DataFrame
summary_df = top_clients_data.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'line_profit': 'sum'
}).reset_index()

# Rename columns for clarity
summary_df.columns = ['Client ID', 'Total Units Purchased', 'Total Shipping Price', 'Total Revenue', 'Total Profit']

# Define the money columns
money_columns = ['Total Shipping Price', 'Total Revenue', 'Total Profit']

# Define a function that converts a dollar amount to millions
def to_millions(amount):
    return amount / 1_000_000

# Apply the function to the money columns
for column in money_columns:
    summary_df[column] = summary_df[column].apply(to_millions)


summary_df['Total Revenue'] = summary_df['Total Revenue'].apply(lambda x: f"${x:,.2f}")
summary_df['Total Shipping Price'] = summary_df['Total Shipping Price'].apply(lambda x: f"${x:,.2f}")
summary_df['Total Profit'] = summary_df['Total Profit'].apply(lambda x: f"${x:,.2f}")
# Rename the columns to reflect the change in the money format
summary_df = summary_df.rename(columns={
    'Total Shipping Price': 'Total Shipping Price (Millions)',
    'Total Revenue': 'Total Revenue (Millions)',
    'Total Profit': 'Total Profit (Millions)'
})
summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_summary_df = summary_df.sort_values(by='Total Profit (Millions)', ascending=False)
sorted_summary_df
```
